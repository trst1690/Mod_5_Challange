## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
#Summary_Stats_byRegimen = duplicate_mouse_cleaned.groupby("Drug Regime")
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
```

```python
#Summary_Stats_byRegimen = duplicate_mouse_cleaned.groupby("Drug Regimen")
sems = duplicate_mouse_cleaned.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
Summary_Stats_byRegimen = duplicate_mouse_cleaned.groupby("Drug Regimen")
tumor_Mean = Summary_Stats_byRegimen["Tumor Volume (mm3)"].mean()
tumor_Median = Summary_Stats_byRegimen["Tumor Volume (mm3)"].median()
tumor_VAR = Summary_Stats_byRegimen["Tumor Volume (mm3)"].var()
tumor_STD = Summary_Stats_byRegimen["Tumor Volume (mm3)"].std()

summary_stats_dict = pd.DataFrame.from_dict({
    "Tumor Mean": tumor_Mean,
    "Tumor Median": tumor_Median,
    "Tumor VAR": tumor_VAR,
    "Tumor STD": tumor_STD,
    "Tumor SEM": sems
})

summary_stats_dict
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
```

```python
treatment_df = duplicate_mouse_cleaned[duplicate_mouse_cleaned['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]
last_timepoint_df = treatment_df.groupby(['Mouse ID'])['Timepoint'].max().reset_index()
tumor_df_merged = last_timepoint_df.merge(treatment_df, on=['Mouse ID', 'Timepoint'], how='left')
final_tumor_volume_df = tumor_df_merged[['Mouse ID', 'Drug Regimen', 'Tumor Volume (mm3)']]
final_tumor_volume_df.head(50)
```

```python
# Put treatments into a list for for loop (and later for plot labels)
#treatment = ["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
#drug_list= []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    #for treatment in drug_list:
        ##tumor_vol = treatment_df['Tumor Volume (mm3)']
    
    # add subset 

    
    # Determine outliers using upper and lower bounds
```

```python
treatment_list = ["Capomulin","Ramicane","Infubinol","Ceftamin"]
drug_list= []
 
for treatment in treatment_list:
    treatment_df = final_tumor_volume_df[final_tumor_volume_df['Drug Regimen'] == treatment]
    tumor_vol = treatment_df['Tumor Volume (mm3)']
    drug_list.append(tumor_vol)
    quartiles = tumor_vol.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
    lower_bound = lower_q - (1.5 * iqr)
    upper_bound = upper_q + (1.5 * iqr)
    outliers = treatment_df.loc[(treatment_df['Tumor Volume (mm3)'] < lower_bound) | 
                                (treatment_df['Tumor Volume (mm3)'] > upper_bound)]
    print(f"Results for {treatment}:")
    print(f"The lower quartile of tumor volume is: {lower_q}")
    print(f"The upper quartile of tumor volume is: {upper_q}")
    print(f"The interquartile range of tumor volume is: {iqr}")
    print(f"The median of tumor volume is: {quartiles[0.5]}")
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    if len(outliers) == 0:
        print("No potential outliers were found.")
    else:
        print(f"{len(outliers)} potential outliers were found:")
        print(outliers)
    print("\n")
        
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
```

```python
Capomulin_regimen_avg_df = Capomulin_regimen_df.groupby(["Mouse ID"]).mean()
x_values = Capomulin_regimen_avg_df['Weight (g)']
y_values = Capomulin_regimen_avg_df['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
